# **Part 3: Detecting and Handling Outliers**

## **Visualizing Outliers Using Box Plots and Histograms**

```python
numerical_columns = ['Age','Fare']
plt.figure(figsize=(10, 6))
sns.boxplot(data=train_df[numerical_columns], palette='Set3')
plt.title('Box Plot to Detect Outliers in Numerical Features')
plt.show()
```

```python
train_df[numerical_columns].hist(bins=30, figsize=(10, 6), color='lightblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

## **Applying statistical methods (e.g., Z-score or IQR) to identify outliers.**

```python

Q1 = train_df[numerical_columns].quantile(0.25)
Q3 = train_df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# To identify Outliers we use this formula Q1 - 1.5*IQR or above Q3 + 1.5*IQR

outliers_iqr = (train_df[numerical_columns] < (Q1 - 1.5 * IQR)) | (train_df[numerical_columns] > (Q3 + 1.5 * IQR))

# Visualizing
plt.figure(figsize=(10, 6))
sns.boxplot(data=train_df[numerical_columns], palette='Set1')
plt.title('Box Plot After IQR Outlier Detection')
plt.show()
```

## **Handling outliers by removing them**

```python

train_df_cleaned = train_df[~((train_df[numerical_columns] < (Q1 - 1.5 * IQR)) |
                              (train_df[numerical_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"Original dataset shape: {train_df.shape}")
print(f"Cleaned dataset shape after removing outliers: {train_df_cleaned.shape}")
```

**After removing outliers, your data will likely follow a more normal or consistent distribution. This helps improve the performance of certain machine learning models, especially those sensitive to skewness or uneven distributions (like linear regression, k-means clustering, etc).**

**There is another method which is called capping.But is we use this method it will create a baisness in model because of artificiall altering value and it will loss it ture nature**

# **Part 7: Feature Scaling**

## **Apply Normalization (Min-Max Scaling)**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_columns = train_df.select_dtypes(include='number').columns
train_df_normalized = train_df.copy()
train_df_normalized[numerical_columns] = scaler.fit_transform(train_df[numerical_columns])

print(train_df_normalized.head())
```

**Why we use it?**

**If our dataset contains feature with different rangs, normalization ensure that all feature have a same scale which is 0 and 1**

**Normalization doesnt assume that the data follows a normal distribution unlike standarization.**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_df_standardized = train_df.copy()
train_df_standardized[numerical_columns] = scaler.fit_transform(train_df[numerical_columns])
print(train_df_standardized.head())
```

## **Visualization: Before and After Scaling (Box Plots and Histograms or Density Plots)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
sns.boxplot(data=train_df[numerical_columns], palette='Set2')
plt.title('Box Plot Before Scaling')

plt.subplot(1, 3, 2)
sns.boxplot(data=train_df_normalized[numerical_columns], palette='Set3')
plt.title('Box Plot After Min-Max Scaling')

plt.subplot(1, 3, 3)
sns.boxplot(data=train_df_standardized[numerical_columns], palette='Set1')
plt.title('Box Plot After Z-score Scaling')

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(18, 12))

plt.subplot(3, 1, 1)
for col in numerical_columns:
    sns.kdeplot(train_df[col], label=col)
plt.title('Density Plot Before Scaling')
plt.legend()

plt.subplot(3, 1, 2)
for col in numerical_columns:
    sns.kdeplot(train_df_normalized[col], label=col)
plt.title('Density Plot After Min-Max Scaling')
plt.legend()


plt.subplot(3, 1, 3)
for col in numerical_columns:
    sns.kdeplot(train_df_standardized[col], label=col)
plt.title('Density Plot After Z-score Scaling')
plt.legend()

plt.tight_layout()
plt.show()
```
